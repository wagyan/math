```
import random
import math
import numpy as np

def isPrime(num):
  k=100
  d=mkd(num)  
  s=int(math.log2(num/d))
  for i in range(k):
    a=random.randint(1,num-1)
    ard=pow(a,d,num)
    if not(ard==1):
      c=0
      for r in range(0, s):
        if not(ard == num-1): c=c+1
        ard=(ard*ard) %num
      if c==s: return False
  return True

def mkd(n):
  d=n-1
  while d&1==0:
    d>>=1
  return d


for a in range(13):
  for b in range(13):
    for c in range(13):
      x=int(str(a+1) + str(b+1) + str(c+1))
      if isPrime(x)==True: print(x)
```
