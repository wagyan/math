```
import sympy
import numpy as np
import math  
from fractions import Fraction

# 楕円曲線上の2点P,Qから、点(P+Q)を求める
def elliptic_group(a,b,c,x1,y1,x2,y2):
  if x1 == x2:
    if y1 == y2 and y1 != 0:
      l = Fraction(3*pow(x1,2) + 2*a*x1 + b, 2*y1  )
    else:
      x3 = x1
      y3 = float('inf')
      return x3,y3
  else:
    l = Fraction(y2-y1,x2-x1)

  n = y1- l*x1
  x3 = l*l - a - x1 - x2
  y3 = - (l*x3 + n)
  return x3,y3

# 三次関数の判別式Dを求める
def calc_D(a,b,c):
  D = -4*pow(a,3)*c + pow(a,2)*pow(b,2) + 18*a*b*c - 4*pow(b,3) - 27*pow(c,2)
  return D

# 因数分解のdictionaryから、2乗数を列挙したリストを返す
def y_squares(dic):
  a = np.array(list( dic.keys()))
  b = np.array(list(dic.values()))
  c = a[b>=2]
  d = b[b>=2]
  e = []
  for i in range(len(c)):
    e.append([pow(c[i],2*x) for x in range(math.floor(d[i]/2)+1) ])

  f = np.array([1])
  for j in range(len(e)):
    g = np.array(e[j])
    f = np.outer(f,g)
    f = f.flatten()
  return list(f)

  
# y^2 = x^3 + a x^2 + b x + c の楕円曲線について
print("y^2 = x^3 + a x^2 + b x + c の楕円曲線の有限位数の有理点を求めます")
sympy.var('x1, a,b,d')
a,b,c = input("スペース込みでa,b,cを入力してください：").split(" ")
a = int(a)
b = int(b)
c = int(c)
sol_lt = []
D = calc_D(a,b,c)
factors = sympy.factorint(D)
print("D={}で因数分解は{}".format(D,factors)) 
y_2 = y_squares(factors)
y_2.append(0)

for i in range(len(y_2)):
  d = c - y_2[i]
  Sol3=sympy.solve (x1**3+a*x1**2+b*x1+d, x1)
  for j in range(len(Sol3)):
    h = complex(Sol3[j])
    if (h.imag == 0):
      h = float(h.real)
      if (h==math.floor(h)):
        sol_lt.append([int(h),int(math.sqrt(y_2[i]))])
        if (y_2[i] != 0):
          sol_lt.append([int(h), - int(math.sqrt(y_2[i]))])

print("整数点の候補は{}".format(sol_lt))

for k in range(len(sol_lt)):
  x = sol_lt[k][0]
  y = sol_lt[k][1]
  print("●P=({},{})の場合".format(x,y))
  x2,y2=x,y
  i=1
  while(y2!=float('inf')):
    i+=1
    x2,y2=elliptic_group(a,b,c,x,y,x2,y2)
    print("{}P=({},{})".format(i,x2,y2))
    if(i>12):break
```
